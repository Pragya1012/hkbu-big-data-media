```python
from bs4 import BeautifulSoup
```

```python
r =requests.get('https://www.oyorooms.com/oyos-in-noida?adults=1&checkin=31%2F03%2F2018&checkout=05%2F04%2F2018&children=0&city=noida&country=India&employee_id=null&guests=1&latitude=null&location=Noida&longitude=null&page=1&rooms=1&searchType=city&src=null')
```

```python
r.text
```

```python
mywebpage = BeautifulSoup(r.text)
```

```python
mywebpage
```

```python
myroom = mywebpage.find_all('span', attrs={'class': 'newHotelCard__hotelName'})
```

```python
myroom
```

```python
rooms = []
for room in myroom:
    rooms.append(room.text.strip())
```

```python
rooms
```

```python
myplace = mywebpage.find_all('span', attrs={'class': 'newHotelCard__hotelAddress'})
```

```python
myplace                   
```

```python
locations = []
for location in myplace:
    locations.append(location.text.strip())
```

```python
locations
```

```python
mystar = mywebpage.find_all('div', attrs={'class': 'hotelRating__dataCapsule'})
```

```python
rating = []
for star in mystar:
    star.find('span').text
    rating.append(star.find('span').text)
```

```python
rating    
```

```python
pragya=[]
for i in range(0,19):
    pragya.append([rooms[i],locations[i],rating[i]])
    
pragya
```

```python
import csv
with open('noidaoyo.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    
    spamwriter.writerows(pragya)
```
